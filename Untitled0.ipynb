```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
plt.scatter(df['wt'],df['mpg'],color='red')
plt.title("relationship between car weight and mpg")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
cylinder=sorted(df['cyl'].unique())
data=[df[df['cyl']==c]['mpg']for c in cylinder]
plt.boxplot(data,tick_labels=cylinder)
plt.title("distribution of mpg")
plt.xlabel("cylinder")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
plt.hist(df['mpg'],color='red',edgecolor='black',bins=2)
plt.title("transmission type is more common[manual,automation]")
plt.xlabel("manual       automation")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv("/content/mtcars (1).csv")
auto_count = (df['am'] == 0).sum()
manual_count = (df['am'] == 1).sum()
labels = ['Automatic', 'Manual']
counts = [auto_count, manual_count]
plt.bar(labels, counts,)
plt.title("transmission Type ")
plt.xlabel(" Type")
plt.ylabel("Number of Cars")
plt.grid(True)
plt.show()
```
