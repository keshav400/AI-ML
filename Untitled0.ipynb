```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
plt.scatter(df['wt'],df['mpg'],color='red')
plt.title("relationship between car weight and mpg")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
cylinder=sorted(df['cyl'].unique())
data=[df[df['cyl']==c]['mpg']for c in cylinder]
plt.boxplot(data,labels=cylinder)
plt.title("distribution of mpg")
plt.xlabel("cylinder")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
plt.hist(df['mpg'],color='red',edgecolor='black')
plt.title("transmission type is more common[manual,automation]")
plt.xlabel("mpg")
plt.ylabel("wt")
plt.grid("True")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
plt.bar(df['wt'],df['mpg'],color=["red","green"])
plt.title("frequency distribution")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.grid("True")
plt.show()
```
